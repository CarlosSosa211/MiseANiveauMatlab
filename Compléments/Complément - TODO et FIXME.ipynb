# Utiliser les éléments TODO et FIXME

Si vous avez suivi les cours précédents, vous avez compris l'importance des commentaires en Matlab. Un aspect intéressant des commentaires est celui de pouvoir ajouter des notes pour la prochaine séance et d'avoir simplement à demander un rapport pour savoir ce qu'il vous reste à faire.

Pour cela, il vous suffit de glisser dans les commentaires les notes :

- TODO : sur ce qu'il vous reste à faire
- FIXME : sur les zones qui posent problèmes
- NOTE : sur les zones qui requièrent votre attention

Prenons par exemple ce fichier : 

```octave
function y = auCarre(x)
%
%   AUCARRE retourne le cube d'un entier
%
%   TODO : Faire la documentation

y = x^3

%   FIXME : Changer le nom de la fonction
%   NOTE : Faire remonter cette coquille au manager
```

Pour faire ressortir ces notes, vous devez utiliser le générateur de rapport de Matlab. Une fois dans le répertoire des fichiers à analyser, cliquez sur la flèche indiquée dans la figure ci-dessous et sélectionnez 'Reports --> TODO/FIXME reports'.




Une fois le rapport fini, vous obtiendrez une image similaire à celle en dessous.



**NB :** Il vous est possible d'utiliser la commande *dofixrpt('.')* pour obtenir le même résultat. L'opération n'étant pas disponible sur Octave, il vous est impossible de tester cette fonctionnalité sur cette plateforme.

*Pour ceux qui l'ont remarqué, il est également possible de faire un 'dependecies report' qui permet de savoir quelles sont les dépendances de nos scripts et fonctions*
